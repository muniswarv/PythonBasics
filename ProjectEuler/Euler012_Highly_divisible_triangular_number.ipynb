### Problem 12: Highly divisible triangular number

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?


```python
import math

def getFactors(n):
    factr_pairs = [ (e, n//e) for e in range(2, int(math.sqrt(n)) + 1  ) if n%e == 0 ]
    factors = [1, n] if not n == 1 else [1]
    for e in factr_pairs: factors.extend(e)
    return sorted(factors)
        
def higlyDivisibleTriangularNumber(n, show=False):
    
    i = 0; num = 0;
    while(True):
        i += 1 ; num += i
        factors = getFactors(num)
        if show: print(num, ":", factors)
        if len(factors) > n: return num
```

```python
higlyDivisibleTriangularNumber(5, show=True)
```

```python
higlyDivisibleTriangularNumber(500)
```

```python
%timeit higlyDivisibleTriangularNumber(500)
```
