#### Q3: Super() Method Override (Super() Override)

```python
class Parent:
    def greet(self):
        return 'Hello from Parent'

class Child(Parent):
    def greet(self):
        res = super().greet()
        return res + ' and Child'
        
p1 = Parent()
print(p1.greet())

c1 = Child()
print(c1.greet())

# Predicted Output: 
# Actual Output: 
```
